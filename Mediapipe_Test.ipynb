```python
cap = cv2.VideoCapture(1)

with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:
    while cap.isOpened():
        # read the raw feed 
        ret, frame = cap.read()
        
        
        # Recolor feed, which needs to be processed by mediapipe 
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # flip the image
        image = cv2.flip(image, 1)
        
#         image.flags.writeable = False
        
        # get the results from mediapipe 
        results = holistic.process(image)
        
#         image.flags.writeable = True

        # recover the color
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Draw face landmarks 
        mp_drawing.draw_landmarks(image, results.face_landmarks, 
                                  mp_holistic.FACEMESH_CONTOURS)
        mp_drawing.draw_landmarks(image, results.pose_landmarks, 
                                  mp_holistic.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=5, circle_radius=5),
                                 mp_drawing.DrawingSpec(color=(255, 105, 180), thickness=5, circle_radius=2))
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, 
                                  mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=5, circle_radius=5),
                                 mp_drawing.DrawingSpec(color=(255, 105, 180), thickness=5, circle_radius=2))
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, 
                                  mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=5, circle_radius=5),
                                 mp_drawing.DrawingSpec(color=(255, 105, 180), thickness=5, circle_radius=2))
        
        # render the imageq
        cv2.imshow("Detection Results", image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break 
```

```python
cap.release()
cv2.destroyAllWindows()
```

```python
# Face Detector 

# BaseOptions = mp.tasks.BaseOptions
# FaceDetector = mp.tasks.vision.FaceDetector
# FaceDetectorOptions = mp.tasks.vision.FaceDetectorOptions
# VisionRunningMode = mp.tasks.vision.RunningMode

# Create a face detector instance with the image mode:
# options = FaceDetectorOptions(
#     base_options=BaseOptions(model_asset_path='/path/to/model.task'),
#     running_mode=VisionRunningMode.IMAGE)
# with FaceDetector.create_from_options(options) as detector:

mp_face_detection = mp.solutions.face_detection 
cap = cv2.VideoCapture(1)

with mp_face_detection.FaceDetection(min_detection_confidence=0.5) as face_detector:
    while cap.isOpened():
        # read the raw feed 
        ret, frame = cap.read()
        if ret is False:
            break


        # Recolor feed, which needs to be processed by mediapipe 
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # flip the image
        image = cv2.flip(image, 1)

        image.flags.writeable = False
        
        # get the shape of the image 
        image_height, image_width, _ = image.shape 

        # get the results from mediapipe 
        result = face_detector.process(image)
        if result.detections:
            for detection in result.detections: 
                bbx = detection.location_data.relative_bounding_box
                
                face_coordinates = np.multiply([bbx.xmin, bbx.ymin, bbx.width, bbx.height], 
                                               [image_width, image_height, image_width, image_height]).astype(int)
                
                start_point = (face_coordinates[0], face_coordinates[1] + face_coordinates[3])
                end_point = (face_coordinates[0] + face_coordinates[2], face_coordinates[1])

                print(start_point)
                print(end_point)
                print("**********")
                
                
                cv2.rectangle(image, start_point, end_point, (255,0,0), 3)
                print(detection)
                print(detection.location_data.relative_bounding_box)
                print(bbx.xmin, bbx.ymin + bbx.height)
                print(bbx.xmin + bbx.width, bbx.ymin)
                print("*********************************************************")
            

        image.flags.writeable = True

        # recover the color
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)


        # render the imageq
        cv2.imshow("Detection Results", image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break 
```
